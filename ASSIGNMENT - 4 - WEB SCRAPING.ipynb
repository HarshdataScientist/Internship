1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome()
```

```python
url = driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Name = []
Artist = []
Date = []
Views = []
```

```python
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]'):
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')
    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]'):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('-')
    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]'):
        Date.append(i.text)
except NoSuchElementException:
    Date.append('-')
    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]'):
        Views.append(i.text)
except NoSuchElementException:
    Views.append('-')
```

```python
Data = pd.DataFrame({})
Data['Name']= Name
Data['Artist']= Artist
Data['Upload Date']= Date
Data['Views']= Views
```

```python
Data.Name = Data.Name.apply(lambda x:x[:-4].strip('"'))
```

```python
driver.close()
```

```python
driver = webdriver.Chrome()
```

```python
url = driver.get('https://archive.ics.uci.edu/')
```

```python
data = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()
```

```python
URL = []
```

```python
URL1 = driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
for i in URL1:
    URL.append(i.get_attribute("href"))
```

```python
Dataset = []
Datatype = []
Task = []
Atype = []
NOI = []
NOA =[]
Year = []
```

```python
for i in URL:
    driver.get(i)
    time.sleep(3)
    
    try:
        dataset = driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        Dataset.append(dataset.text)
    except NoSuchElementException:
        Dataset.append('-')
    time.sleep(3)
        
    try:
        datatype = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        Datatype.append(datatype.text)
    except NoSuchElementException:
        Datatype.append('-')
    time.sleep(3)
        
    try:
        task = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
    time.sleep(3)
        
    try:
        atype = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        Atype.append(atype.text)
    except NoSuchElementException:
        Atype.append('-')
    time.sleep(3)
        
    try:
        noi = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        NOI.append(noi.text)
    except NoSuchElementException:
        NOI.append('-')
    time.sleep(3)
        
    try:
        noa = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        NOA.append(noa.text)
    except NoSuchElementException:
        NOA.append('-')
    time.sleep(3)
        
    try:
        year = driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text)
    except NoSuchElementException:
        Year.append('-')
    time.sleep(3)
```

```python
len(Dataset),len(Datatype),len(Task),len(Atype),len(NOI),len(NOA),len(Year)
```

```python
df = ({'Dataset Name': Dataset,
      'Data type': Datatype,
      'Task': Task,
      'Attribute Type': Atype,
      'No Of Instances': NOI,
      'No Of Attribute': NOA,
      'Year': Year})
```
